```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def read_tsv(file_path):
    """Læs en TSV-fil og returner x- og y-værdier."""
    data = np.loadtxt(file_path, delimiter='\t')
    x = data[:, 0]
    y = data[:, 1]
    return x, y

def analyze_and_plot(x, y, title, ax):
    """Beregner statistikker, lineær regression og plotter data."""
    # Gennemsnit
    mean_x, mean_y = np.mean(x), np.mean(y)

    # Varians
    var_x, var_y = np.var(x), np.var(y)

    # Pearsons korrelation
    corr = np.corrcoef(x, y)[0, 1]

    # Lineær regression
    slope, intercept, _, _, _ = stats.linregress(x, y)

    # Print resultater
    print(f"{title} - Gennemsnit: x = {mean_x:.2f}, y = {mean_y:.2f}")
    print(f"{title} - Varians: x = {var_x:.3f}, y = {var_y:.3f}")
    print(f"{title} - Pearsons korrelation: {corr:.3f}")
    print(f"{title} - Lineær regression: f(x) = {slope:.2f}x + {intercept:.2f}\n")

    # Plot data og regression
    ax.scatter(x, y, color='blue')
    ax.plot(x, slope * x + intercept, color='red')
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

# Læs datafilerne
x1, y1 = read_tsv('data1.tsv')
x2, y2 = read_tsv('data2.tsv')
x3, y3 = read_tsv('data3.tsv')
x4, y4 = read_tsv('data4.tsv')

# Opret subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Analyser og plot data
analyze_and_plot(x1, y1, 'Data 1', axs[0, 0])
analyze_and_plot(x2, y2, 'Data 2', axs[0, 1])
analyze_and_plot(x3, y3, 'Data 3', axs[1, 0])
analyze_and_plot(x4, y4, 'Data 4', axs[1, 1])

# Juster layout
plt.tight_layout()
plt.show()
```
