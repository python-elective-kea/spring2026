```python
  # imports
  from dotenv import load_dotenv
  import os
```

```python
import requests

load_dotenv()
api_key = os.environ.get('API_KEY')
url = 'https://api.github.com/repos/clbokea/osman'

headers = {
    'Authorization': f'token {api_key}'
}
response = requests.get(url, headers=headers)
repo_data = response.json()
print(repo_data)
```
