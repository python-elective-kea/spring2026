## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 0])
y = [100, 12, 14, 16]

plt.bar(x, y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python

x = np.linspace(0, 200, 400)

y1 = x
y2 = x/2
y3 = np.log2(x)

plt.plot(x, y1, label='n', color='red')
plt.plot(x, y2, label='n/2', color='blue')
plt.plot(x, y3, label='log2(n)', color='black')

plt.title('Big o notation')
plt.legend()

plt.ylim(0,12)
plt.xlim(0,200)

plt.xlabel('Input size(n)')
plt.ylabel('Time complexity')

plt.grid(True)
plt.show()
```
