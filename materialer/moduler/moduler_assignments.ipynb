```python
# download_page_exam.py
import os
import sys
import requests
import subprocess

# 1️⃣ Opret mappe
folder = "downloaded_site"
os.makedirs(folder, exist_ok=True)

# 2️⃣ Download HTML
html_url = "https://itakea.github.io/e24_swa/py_intro_3.html"
html_file = os.path.join(folder, "py_intro_3.html")
html_content = requests.get(html_url).text

# 3️⃣ Download stylesheets og ændr links
lines, new_lines = html_content.splitlines(), []
for line in lines:
    if '<link' in line and 'stylesheet' in line and 'href="' in line:
        start = line.find('href="') + 6
        end = line.find('"', start)
        css_url = line[start:end]
        
        # Hvis CSS-URLen er relativ, lav den absolut
        if not css_url.startswith('http'):
            css_url = html_url.rsplit('/', 1)[0] + '/' + css_url
        
        css_name = os.path.basename(css_url.split("?")[0])
        css_path = os.path.join(folder, css_name)
        with requests.get(css_url) as r:
            with open(css_path, "w", encoding="utf-8") as f:
                f.write(r.text)
        new_lines.append(line[:start] + css_name + line[end:])
    else:
        new_lines.append(line)

# 4️⃣ Gem HTML med lokale CSS-links
with open(html_file, "w", encoding="utf-8") as f:
    f.write("\n".join(new_lines))

# 5️⃣ Åbn i browser
if sys.platform.startswith("win"):
    subprocess.run(["start", html_file], shell=True)
elif sys.platform == "darwin":
    subprocess.run(["open", html_file])
else:
    subprocess.run(["xdg-open", html_file])
```
